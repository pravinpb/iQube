```python
out = []

m = np.mean(data["SQUARE_FT"])

s = np.std(data["SQUARE_FT"])

threshold = 1

list_index = list(data["SQUARE_FT"])

for i in data["SQUARE_FT"]:
    r = (i - m ) / s
    r = np.abs(r)
    if threshold < r :
        
        out.append(list_index.index(i))
out

for j in out:
    data.at[j,"SQUARE_FT"] = round(m)
```
